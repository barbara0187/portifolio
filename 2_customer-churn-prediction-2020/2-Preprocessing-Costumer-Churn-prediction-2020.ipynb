# Data preprocessing

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import pylab as pl
import numpy as np
from scipy import stats as stats

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None) 

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
data = pd.read_csv('train.csv')
data.head()
```

Let's eliminate some of the variables that we find that are eequivalent in the last notebook.

```python
data.drop(['total_day_charge', 'total_eve_charge', 'total_night_charge', 'total_intl_charge'], axis=1, inplace=True)
data.head()
```

## Codificação de variáveis categoricas nominais

In this notebook we have five categorical variables. We need to change the categories of international_plan, voice_mail_plan and churn to numbers, we do that by mapping yes to 1 and no to zero. For the variables state and area_code we have many categories, than we use the get_dummies to criate new variables labeled with 0 or 1.

```python
data['international_plan'] = data['international_plan'].map({'yes':1, 'no':0})
data['voice_mail_plan'] = data['voice_mail_plan'].map({'yes':1, 'no':0})
data['churn'] = data['churn'].map({'yes':1, 'no':0})
```

```python
data_dummies = pd.get_dummies(data[['state','area_code']], drop_first=True, dtype=int)
data_dummies.head()
```

### Criate the final dataset

Here we concatenate our data (without state and area_code) with data_dummies in order to produce a final dataframe.

```python
final_data=pd.concat([data.drop(['state','area_code'], axis=1),data_dummies], axis=1)
final_data.head()
```

## Definindo X e y

We define who are our dependent variables in dataframe X and our independent variable in the series y.

```python
X = final_data.drop(['churn'], axis=1)
y =final_data.churn
```

## Separação de bases

We alread had a test dataset that Kaggle offers, althoug we have no way to check if the answers of this dataset because the compeition is alread closed. Then, we criate our on train and test sets from the final_data we had produced.

```python
from sklearn.model_selection import train_test_split

y_train, y_X_train, X_test, test =train_test_split(X, y, test_size = .2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Feature Scaling

Here we normalize the train dataset to be between 0 and 1. then we fit this normalization in the test set.

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

colunas = X_train.columns

X_train[colunas] = scaler.fit_transform(X_train)
X_test[colunas] = scaler.transform(X_test)
```

# Exporting the data

Finally its time to export our data to be used in the machine learning models.

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv',index = False)
y_test.to_csv('y_test.csv',index = False)
```
