```python
sns.countplot(data=data, x='churn')
```

```python
data['international_plan'] = data['international_plan'].map({'yes':1, 'no':0})
data['voice_mail_plan'] = data['voice_mail_plan'].map({'yes':1, 'no':0})
data['churn'] = data['churn'].map({'yes':1, 'no':0})
```

```python
len(data.churn[data['churn']==1])/len(data.churn)
```

# Finding paterns with pairplot

From the analises of pairplot we find variables that are equvalent: total_day_minutes and total_day_charge; total_eve_minutes and total_eve_charge; total_night_minutes and total_night_charge; total_intl_minutes and total_intl_charge. In the next notebook we need to eliminate one variable of each pair.

```python
data.columns
```

```python
data0=data[['state', 'account_length', 'area_code', 'international_plan',
       'voice_mail_plan', 'churn']]
sns.pairplot(data= data0)
```

```python
data1=data[['number_vmail_messages', 'total_day_minutes',
       'total_day_calls', 'total_day_charge', 'churn']]
sns.pairplot(data= data1)
```

```python
data2=data[['total_eve_minutes', 'total_eve_calls', 'total_eve_charge', 'total_night_minutes',
       'total_night_calls', 'total_night_charge', 'churn']]
sns.pairplot(data= data2)
```

```python
data3=data[['total_intl_minutes',
       'total_intl_calls', 'total_intl_charge',
       'number_customer_service_calls', 'churn']]
sns.pairplot(data=data3)
```
